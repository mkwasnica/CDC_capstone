```python
# create a function that iters thru rows and look for NA's (tally) and if > 23, drop that row

def check_NA_per_row(df):
    num_of_na_per_row = []
    idx_to_drop = []
    for i in range(len(df)):
        num_of_na_per_row += [df.iloc[i].isnull().sum()]
#         if i % 100000 == 0:
#             print(i)
    #should i return the indicies of where to drop, or the df with those rows dropped?
    for idx in range(len(num_of_na_per_row)):
        if num_of_na_per_row[idx] > 12:
            idx_to_drop += [idx]
    return idx_to_drop
```
